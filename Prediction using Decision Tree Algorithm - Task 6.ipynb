### üìå Shuffling the dataset

```python
iris_data = iris_data.sample(frac=1)
iris_data.head(10)
```

###### üí†Observation: Comparing the last columns, the labels are:üí†
* Iris-setosa - 0
* Iris-versicolor - 1
* Iris-virginica - 2
 

### üìå Encoding the Target Data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encode = LabelEncoder()
iris_data['Encoded Species'] = label_encode.fit_transform(iris_data['Species'])
iris_data.head(149)
```

```python
# Converting label column into an array
enc_data = label_encode.fit_transform(iris_data['Species'])
enc_data
```

### üìå Splitting the Data

```python
from sklearn.model_selection import train_test_split
X = iris_data.iloc[:,:-1].values
X
```

```python
#New shape
X.shape 
```

```python
Y = enc_data
Y
```

```python
# Splitting data for Training (70%) & Testing (30%) 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

### üìå Building the Decision Tree Model, Training it and Doing Predictions

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier() # creating an instance
dtree.fit(X_train, y_train) # fitting X_train & y_train into Decision Tree
```

```python
predictions = dtree.predict(X_test)
# A Classification report is used to measure the quality of predictions from a classification algorithm.
# A Confusion matrix is a summary of prediction results on a classification problem.
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, predictions))
```

###### üí†Observation: The model collects an accuracy of about 96%üí†

```python
print(confusion_matrix(y_test, predictions))
```

### üìå Testing our model on a random test case

```python
# Predicted label should be Iris-virginica
trial = np.array([[7.2,3.0,5.8,1.6]]) 
pred_trial = dtree.predict(trial)
pred_trial
```

###### üí†Observation: Array([2]) or Label 2 belongs to Iris-virginica, hence the prediction is right.üí†
